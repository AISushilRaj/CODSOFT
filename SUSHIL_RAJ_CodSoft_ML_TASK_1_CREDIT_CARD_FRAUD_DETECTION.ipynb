```python
filename=list(uploaded.keys())[0]
```

```python
import pandas as pd
df=pd.read_excel(filename)
df.head()
```

```python
X = df.drop(columns=["is_fraud"])
y = df["is_fraud"]
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split


datetime_cols = X.select_dtypes(include=["datetime64"]).columns.tolist()


for col in datetime_cols:
    X[col] = X[col].astype("int64") // 10**9


categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()


X = pd.get_dummies(X, columns=categorical_cols)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)


dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)


y_pred = dt_classifier.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

LogisticRegression

```python
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)

y_pred_lr = lr_classifier.predict(X_test)


print("Logistic Regression","Accuracy Score:", accuracy_score(y_test, y_pred_lr))
print("Logistic Regression",classification_report(y_test, y_pred_lr))
```

RandomForestClassifier

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

y_pred_rf = rf_classifier.predict(X_test)

print("\nRandom Forest Model:")
print("Accuracy Score:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

 # Train again after balancing

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

rf_classifier.fit(X_train_smote, y_train_smote)

y_pred_rf_smote = rf_classifier.predict(X_test)

print("\nRandom Forest with SMOTE:")
print("Accuracy Score:", accuracy_score(y_test, y_pred_rf_smote))
print(classification_report(y_test, y_pred_rf_smote))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_smote, y_train_smote)

print(" Best Hyperparameters:", grid_search.best_params_)
#.....................................................................for better model
best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test)

print("\n Best Random Forest Model:")
print("Accuracy Score:", accuracy_score(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))
```
